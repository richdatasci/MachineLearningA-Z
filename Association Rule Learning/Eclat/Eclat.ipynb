```python
# import the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import dataset
# the dataset doesn't have a header,  so the header argument needs to be set
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
# create a list of transations 
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i, j]) for j in range(0, 20)])
```

```python
# training the eclat mode on the dataset
from apyori import apriori
# min support is 3 products, times 7 days, divided by 7501 transactions
# trialing different confidence levels with this dataset indicated 0.2 was optimal
# minimum lift of 3 will give relevant rules.
# min_leng and max_leng to reflect business requirements.
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2,
                min_lift = 3, min_length = 2, max_length = 2)
```

```python
# display the result coming directly from the output of the model
results = list(rules)
results
```

```python
# put the results into a pandas dataframe
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```
